## Bar Chart

Bar charts are a common type of data visualization used to represent categorical data. They consist of rectangular bars, where the length or height of each bar is proportional to the value it represents. 

```python
import numpy as np
```

```python
company = ['Google', 'Amazon', 'Microsoft', 'Facebook']
revenue = [90, 136, 89, 27]
```

```python
plt.bar(company, revenue)
```

```python
#Use numpy to change string x labes into integers
xpos = np.arange(len(company))
xpos
```

```python
plt.bar(xpos, revenue)
```

```python
#To label the xpos with corresponding string values have to use 'xticks' method
plt.xticks(xpos, company)
plt.bar(xpos, revenue)
```

Plot two things on the same chart

```python
profit = [40, 2, 34, 12]
plt.bar(xpos, revenue, label = 'Revenue')
plt.bar(xpos, profit, label = 'Profit')

plt.legend()
```

```python
plt.bar(xpos-0.2, revenue, label = 'Revenue', width = 0.4)
plt.bar(xpos+0.2, profit, label = 'Profit', width = 0.4)

plt.legend()
```

Horizontal bar chart

```python
plt.yticks(xpos, company)
plt.barh(xpos-0.2, revenue, label = 'Revenue', height = 0.4)
plt.barh(xpos+0.2, profit, label = 'Profit', height = 0.4)
plt.legend()
```

## Histogram

A histogram is a graphical representation of the distribution of a dataset. It's a type of bar chart that displays the frequency or count of data points in specific intervals or "bins."

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]

plt.hist(blood_sugar)
```

```python
plt.hist(blood_sugar, bins = 3, rwidth = 0.95)

# array([ 77.        , 101.33333333, 125.66666667, 150.        ]
```

```python
plt.hist(blood_sugar, bins = [80, 100, 125, 150], rwidth = 0.95)

# array([ 80, 100, 125, 150]
```

```python
plt.hist(blood_sugar, bins = [80, 100, 125, 150], histtype = 'step')
```

```python
plt.hist(blood_sugar, bins = [80, 100, 125, 150], rwidth = 0.95, orientation = 'horizontal')
```

Plot two types of hist side by side

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]

plt.hist([blood_sugar_men, blood_sugar_women], bins = [80, 100, 125, 150], rwidth = 0.95, color = ['g', 'orange'], label = ['men', 'women'])

plt.legend()
```

## Pie chart

A pie chart is a circular graphical representation of data, where the whole circle represents 100% of the data, and each segment (or "slice") of the pie represents a portion of that whole, typically expressed as a percentage.

```python
exp_vals = [1400, 600, 300, 410, 250]
exp_labels = ["Home Rent", "Food", "Phone bill", "Car", "Other"]

plt.pie(exp_vals, labels=exp_labels)
plt.show()
```

```python
plt.pie(exp_vals, labels=exp_labels, autopct = '%.2f%%')
plt.show()
```

```python
plt.pie(exp_vals, labels=exp_labels, autopct = '%.2f%%', explode = [0,0.2,0,0,0.2])
plt.show()
```

```python
plt.pie(exp_vals, labels=exp_labels, autopct = '%.2f%%', explode = [0,0.2,0,0,0.2], startangle = 45)

plt.savefig("pie.jpg", bbox_inches = 'tight')
plt.show()
```
